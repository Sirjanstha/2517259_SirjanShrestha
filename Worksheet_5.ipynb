```python
#task-4
# building a cost function
import numpy as np

def cost_function(X, Y, W):
    """
    Parameters:
    This function finds the Mean Square Error.

    Input parameters:
    X : Feature Matrix (m × n)
    Y : Target Vector (m × 1) or (m,)
    W : Weight Vector (n × 1)

    Output:
    cost : Mean Squared Error
    """

    m = len(Y)  # number of training examples

    # Predictions
    Y_pred = X @ W

    # Mean Squared Error
    cost = (1 / (2 * m)) * np.sum((Y_pred - Y) ** 2)

    return cost
```

```python
def gradient_descent(X, Y, W, alpha=0.01, iterations=1000):
    m, n = X.shape
    W = np.zeros(n)
    cost_history = []

    for _ in range(iterations):
        # Prediction
        Y_pred = X @ W

        # Loss
        loss = Y_pred - Y

        # Cost (Mean Squared Error)
        cost = (1 / (2 * m)) * np.sum(loss ** 2)
        cost_history.append(cost)

        # Gradient
        dw = (1 / m) * (X.T @ loss)

        # Update weights
        W = W - alpha * dw

    return W, cost_history
```

```python
import numpy as np

# Model Evaluation - RMSE
def rmse(Y, Y_pred):
    """
    This Function calculates the Root Mean Squared Error.
    Input Arguments:
    Y: Array of actual (target) dependent variables
    Y_pred: Array of predicted dependent variables
    Output:
    rmse: Root Mean Square Error
    """
    rmse = np.sqrt(np.mean((Y_pred - Y) ** 2))
    return rmse
```

```python
#9
import numpy as np

def r2(Y, Y_pred):
    """
    This Function calculates the R Squared Error.

    Input Arguments:
    Y      : Array of actual (target) dependent variables
    Y_pred : Array of predicted dependent variables

    Output Arguments:
    r2 : R Squared Error
    """

    mean_y = np.mean(Y)

    # Total Sum of Squares
    ss_tot = np.sum((Y - mean_y) ** 2)

    # Residual Sum of Squares
    ss_res = np.sum((Y - Y_pred) ** 2)

    # R-squared
    r2 = 1 - (ss_res / ss_tot)

    return r2
```
